```python
#Function cleaning
```

```python
def height(text):
    return text.replace("'", ".").replace('"' , '')

def stars(text):
    return text[0]

def money(text):
    return text.replace("€", "").replace("K", "").replace("M", "")

def weight(text):
    return text.replace("lbs", "")

def add_values(text):
    return sum([int(x) for x in text.split('+')])
    
def clean_fifa21(df):
    df2 = df.copy()
    #Uniform columns names
    df2.columns= df2.columns.str.lower()
 
    #Clean numerical columns
    df2_numerical = df2.select_dtypes(include=np.number).copy()

    for column in ['volleys', 'curve', 'agility', 'balance', 'jumping','interceptions','positioning','vision','composure','sliding tackle']:
        mean = df2_numerical[column].mean()
        df2_numerical[column] = df2_numerical[column].fillna(mean)
        
    #Clean categorical columns
    df2_categorical = df2.select_dtypes(exclude=np.number).copy()
    for column in ['club', 'position', 'joined', 'a/w', 'd/w']:
        mode = df2_categorical[column].mode()
        df2_categorical[column] = df2_categorical[column].fillna(mode[0])
        
    df2_categorical = df2_categorical.drop(['loan date end'], axis=1) # axis = 1 -> columns

    for col in ['height']:
        df2_categorical[col] = df2_categorical[col].apply(height)
    
    for col in ['w/f','sm', 'ir']:
        df2_categorical[col] = df2_categorical[col].apply(stars)
    
    for col in ['value','wage', 'release clause','hits']:
        df2_categorical[col] = df2_categorical[col].apply(money)
    
    for col in ['weight']:
        df2_categorical[col] = df2_categorical[col].apply(weight)
    
    df2_categorical_l=df2_categorical.iloc[:,:20]
    df2_categorical_r=df2_categorical.iloc[:,21:]
    for col in df2_categorical_r.columns:
        df2_categorical_r[col] = df2_categorical_r[col].apply(add_values)
        
    #Concatenate 
    df2= pd.concat([df2_categorical_l,df2_categorical_r,df2_numerical], axis=1)
   
    return df2
```
